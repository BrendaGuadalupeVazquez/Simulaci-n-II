Acierto y error

```python
import matplotlib.pyplot as plt #libreria para graficar
import numpy as np
import random as rd #libreria para generar numeros aleatorios
```

```python
def f(x):
  return np.sqrt(np.arctan(x))#la funcion de la cual queremos graficar y sproximar su integral
```

```python
aciertos=0 #numero de puntitos bajo la curva
n=1000 #numero de puntitos
```

```python
#generadores de puntos x & y aleatorios
x=rd.random()
y=rd.random()
```

```python
a=np.linspace(0,1,100)
b=f(a)
```

```python
plt.plot(a,b,color='blue')
plt.scatter(x,y,color='green')
```

```python
for i in range(1000):
  x=rd.random()
  y=rd.random()
  plt.plot(a,b,color='blue')
  plt.scatter(x,y,color='green')
```

```python
for i in range(n):
  x=rd.random()
  y=rd.random()
  plt.plot(a,b,color='red')
  if y<=f(x):
    plt.scatter(x,y,color='green')
    aciertos+=1
  else: plt.scatter(x,y,color='yellow')
```

```python
I=aciertos/n
```
