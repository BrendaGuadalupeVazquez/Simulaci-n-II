#**LÍNEA DE ESPERA CON UN SERVIDOR**

```python
import numpy
import matplotlib.pyplot as plt
from random import expovariate
from statistics import mean
from math import inf
```

```python
lamda=4.0
mu=6.0
Num_Pkst=10000
count=0
t=0
N=0
tll=expovariate(lamda)
ts=inf
t_evento_anterior=0.0
tll_Data=[]
ts_Data=[]
r_Data=[]
area=[]
Data={}
cola_llegadas=[]
while count<Num_Pkst:
    if tll<ts:
        t=tll
        delta=t-t_evento_anterior
        if N in Data:
          Data[N]+=delta
        else:
          Data[N]=delta
        area.append(delta*N)
        t_evento_anterior=t
        N=N+1.0
        cola_llegadas.append(t)
        tll_Data.append(t)
        tll=t+expovariate(lamda)
        if N==1:
            ts=t+expovariate(mu)
    else:
        t=ts
        delta=t-t_evento_anterior
        if N in Data:
          Data[N]+=delta
        else:
          Data[N]=delta
        area.append(delta*N)
        t_evento_anterior=t
        N=N-1.0
        t_llegada=cola_llegadas.pop(0)
        ts_Data.append(t)
        r_Data.append(t-t_llegada)
        count=count+1
        if N>0:
            ts=t+expovariate(mu)
        else:
            ts=inf
print("Retraso promedio = W = ",round(mean(r_Data),4))
for (key,value) in Data.items():
    Data[key]=value/t
print("Suma de probabilidades=",sum(Data.values()))
media=0.0
for (key,value) in Data.items():
    media=media+key*value
print("E[N]=",media)
print("\nProbabilidades P(N=k):")
for k in sorted(Data.keys()):
    print("P(N=",k,")=",round(Data[k],4))
```

Para los parámetros $\lambda = 4$ y $\mu = 6$, se obtiene:

$\rho = \frac{4}{6} = \frac{2}{3}≈0.6667$

$P(N = k) = \frac{1}{3}\left(\frac{2}{3}\right)^k$

$E[N] = 2$

$W = 0.5$

$W_q = \frac{1}{3}≈0.3333$

$L_q = \frac{4}{3}≈1.3333$
